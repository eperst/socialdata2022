```python
import pandas as pd

url = 'https://data.melbourne.vic.gov.au/api/views/fp38-wiyy/rows.csv?accessType=DOWNLOAD'
trees_data = pd.read_csv(url)
trees_data.head()
```

```python
cols = ['Suburb', 'Address', 'Rooms', 'Type', 'Price', 'Method', 'SellerG',
       'Date', 'Distance', 'Postcode', 'Bedroom2', 'Bathroom', 'Car',
       'Landsize', 'BuildingArea', 'YearBuilt', 'CouncilArea', 'Latitude',
       'Longitude', 'Regionname', 'Propertycount']
housing_data = pd.read_csv("../../../../Downloads/Melbourne_housing_FULL.csv", header=0, names=cols)
housing_data.head()
```

```python
house_placements = pd.DataFrame(housing_data[["Latitude", "Longitude"]])
house_placements.dropna(inplace=True)
house_placements.head()
```

```python
import shapefile
from Shapely.geometry import Point
from Shapely.geometry import shape

sf = shapefile.Reader("../../../../Downloads/Municipal boundary/melbourne.shp")
for boundary in sf.shapes():
    if Point((-37.5, 145)).within(shape(boundary)):
        print("contains")
```

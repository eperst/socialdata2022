```python
import pandas as pd

df = pd.read_csv('../files/assignment_1_feedback.csv')
df.head()
```

```python
for c in df.columns:
    df[c]
    print(f"{c}:\nMin: {df[c].min()}\nMax: {df[c].max()}")
```

```python
ours = df.loc[26]
all_grades = len(df)
ours
```

```python
for idx, c in enumerate(df.columns):
    worse = len(df[c].where(lambda x : x < ours[idx]).dropna())
    print(f"{c} better than {worse:.2f}% of groups")
   
```

```python
import matplotlib.pyplot as plt
import numpy as np
b = [0.00, 0.25, 0.50, 0.75, 1.00, 1.25, 1.50, 1.75, 2.00, 2.25, 2.50, 2.75, 3.00]
fig, axs = plt.subplots(2, 3, constrained_layout=True, figsize=(16, 10))
intervals = [pd.Interval(left=left, right=b[idx + 1], closed='right') for idx, left in enumerate(b[:-1])]
cats = [f"{i.left:.2f}-{i.right:.2f}" for i in intervals]
for idx, c in enumerate(df.columns):
    [cat] = pd.cut([ours[c]], bins=b, right=True)
    grade_categories = pd.cut(df[c], bins=b, right=True)
    x = idx % 3
    y = idx // 3
    ratio = df[c].groupby(grade_categories).size() / len(grade_categories)
    bars = axs[y, x].barh(range(len(ratio.index)), ratio.values, height=0.3, tick_label=cats)
    our_interval = intervals.index(cat)
    bars[our_interval].set_color('r')
    axs[y, x].set_title(c)
plt.suptitle("Ratio of groups in each score range for Assignment 1")
plt.show()
```

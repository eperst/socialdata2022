```python
# Initial setup
import pandas as pd

# Read csv data
data = pd.read_csv("Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv")

# Display sample data
data.head()
```

```python
# Excercise 1: computing simple statistics

categories = data["Category"].unique()

crime_no_per_category = {c : len(data[data["Category"] == c]) for c in categories }

print("Total number of crimes: {total_crimes}".format(total_crimes=data.shape[0]))
print("Distinct crime categories:")
print(*categories, sep=', ')
print("Number of crime categories: {0}".format(len(categories)))
print("Number of crimes per category:")
for k in crime_no_per_category.keys():
    print("{0}: {1}".format(k, crime_no_per_category[k]))

```

```python
# Excercise 2: most frequent and least frequent crimes
categorised_crimes = pd.Series(crime_no_per_category)
print(
"""Most common category of crime: {0}
Least common category of crime: {1}"""
.format(categorised_crimes.idxmax(), categorised_crimes.idxmin()))
```

```python
# Excercise 2: plotting crime occurrences
import matplotlib.pyplot as plt

ax = categorised_crimes.plot.bar(x="Categories", figsize=(14, 7), fontsize=12)
ax.set_title("Crime occurrences by category", fontsize=16)
ax.set_ylabel("Occurrences", fontsize=16)
ax.set_xlabel("Category", fontsize=16)
plt.tight_layout()
plt.show()


```

**Records of 2018 are incomplete**

The last recorded date of incident is 2018-05-15, which suggests that the data is incomplete.
Yearly statistics should exclude data from 2018.

```python
data["Date"] = pd.to_datetime(data["Date"], format="%m/%d/%Y")
crimes_2018 = data.loc[(data["Date"] >= "2018-01-01") & (data["Date"] < "2019-01-01")]
crimes_2018["Date"].max()
```

```python
#Exercise 3: Crimes per year, excluding 2018

data["Date"] = pd.to_datetime(data["Date"], format="%m/%d/%Y")
complete_years = data.loc[(data["Date"] >= "2003-01-01") & (data["Date"] < "2018-01-01")]
crime_by_years = complete_years.groupby(pd.Grouper(key="Date", freq="1Y")).size()

print("""Year with the least crimes: {min}
Year with the most crimes: {max}
Average number of crimes per year: {avg}""".format(min=crime_by_years.idxmin().year, max=crime_by_years.idxmax().year, avg=crime_by_years.mean()))
```

```python
# Exercise 3: plotting the average number of crimes per year by category with standard deviation
averaged = complete_years.groupby(["Category", "Year"]).size().unstack("Category", fill_value=0).mean()
averaged["std"] = averaged.std()
ax = averaged.plot(kind="bar", x="",y="mean", xerr="std", figsize=(14, 7), fontsize=12)
ax.set_title("Average number of crimes per year by category", fontsize=16)
ax.set_ylabel("Occurrences", fontsize=16)
ax.set_xlabel("Category", fontsize=16)
plt.tight_layout()
plt.show()

print(averaged)
```

```python
# Exercise 3: plotting number of focus crimes per year

focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])

focused_data = complete_years[complete_years["Category"].isin(focuscrimes)]
grouped = focused_data.groupby(["Category", "Year"]).size().unstack("Category", fill_value=0)
fig = grouped.plot(kind="bar", subplots=True, figsize=(16, 26), layout=(7, 2), legend=False, title="Number of crimes per year by category")

plt.show()
```

```python
import pandas as pd
data_f = pd.read_csv('../files/data9b_f.csv')
data_m = pd.read_csv('../files/data9b_m.csv')
```

```python
import math

opt_bin_f = math.ceil(math.log(len(data_f)) + 1)
opt_bin_m = math.ceil(math.log(len(data_m)) + 1)
print(f"Opt bin-size: Women {opt_bin_f}, Men {opt_bin_m}")
```

```python
mean_steps_f = data_f["steps"].mean()
mean_steps_m = data_m["steps"].mean()
print(f"""Mean steps:
Women: {mean_steps_f}, Men: {mean_steps_m}
Difference: {mean_steps_m - mean_steps_f}""")
```

```python
from scipy import stats

stats.ttest_ind(data_f["steps"], data_m["steps"])
```

```python
import matplotlib.pyplot as plt
```

```python
#Ex 1.2 / 1: histograms of men's and women's daily step count
fig, ax = plt.subplots(2, 1, figsize=(10, 8))
plt.suptitle("Step count distributions for men and women")
ax[0].hist(data_f["steps"], bins=opt_bin_f, rwidth=0.9)
ax[0].set_title("WOMEN")
ax[1].hist(data_m["steps"], bins=opt_bin_m, rwidth=0.9)
ax[1].set_title("MEN")
plt.show()
```

No statistically significant difference between men and women concerning daily step count.
H0 was rejected.

```python
def pearson(dataset, xkey, ykey):
    std = dataset.std()
    covariance = dataset.cov()[xkey][1]
    return covariance / std[xkey] * std[ykey]
```

```python
# Ex 1.2 / 2: Pearson correlation between steps and BMI in men and women

p_f = pearson(data_f, "steps", "bmi")    
p_m = pearson(data_m, "steps", "bmi")

print(f"Men:  {p_m:10.3f}")    
print(f"Women:{p_f:10.3f}")
```

Both H2 and H3 are verified. There is a more significant inverse correlation for women than for men (over twice as much).

#### 1
The jitter plot visualizes the values of a dataset while ensuring that the resulting points don't fall on top of each other. This is achieved by offsetting the points by a small distance vertically and/or horizontally.

#### 2
Figure 2-3 shows how the choice of bin anchoring points can significantly change the appearance of a histogram. 

#### 3
KDE's can be misleading because our eyes are better at comparing distances than areas. This can lead to the peak of the KDE appearing more significant than it is with respect to the whole dataset.

#### 4
It is easier to understand the real world implications of a KDE than those of a CDF. Specific numbers of occurrences are easier to grasp than cumulative normalized data.

#### 5
A **Quantile plot** is a different way to display a CDF. It shows the value (on the y axis) corresponding to a chosen percentile (on the x axis) of the dataset.

#### 6
A **Probability plot** displays the values of the dataset as a function of their respective quantiles in order to compare their distribution to a known distribution (e.g. Gaussian). The function used comes from the inverse theoretical distribution we are comparing with. 
I have not seen this type of plot before.

#### 7
For summary statistics to return meaningful information about the data, some prerequisites must be fulfilled (e.g. a mean applies to unimodal distributions). Unless the data fulfills the prerequisites, summary statistics can be misleading.

#### 8
Box plots are most useful when comparing different distributions' main characteristics. It is easy to spot the similarities and differences at a glance.

#### 9
Violin plots are more informative than box plots (as they include the KDE as well), but are harder to understand and compare.

> * Let's make a jitter-plot (that is, code up something like **Figure 2-1** from DAOST from scratch), but based on *SF Police data*. My hunch from inspecting the file is that the police-folks might be a little bit lazy in noting down the **exact** time down to the second. So choose a crime-type and a suitable time interval (somewhere between a month and 6 months depending on the crime-type) and create a jitter plot of the arrest times during a single hour (like 13-14, for example). So let time run on the $x$-axis and create vertical jitter.

```python
import pandas as pd
data = pd.read_csv("Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv")
data.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def jitter(dots):
    offsets = dots.get_offsets()
    offsets[:, 1] += np.round(np.random.uniform(-0.4, 0.4, offsets.shape[0]), decimals=2)
    dots.set_offsets(offsets)
    
    
```

```python
import datetime

data["Date"] = pd.to_datetime(data["Date"], format="%m/%d/%Y")
data["Time"] = pd.to_datetime(data["Time"])
assaults_jan_2017 = data.loc[(data["Category"] == "ASSAULT") & (data["Date"] >= '2017-01-01') & (data["Date"] <= '2017-01-31')]
pdata = assaults_jan_2017.loc[assaults_jan_2017.Time.dt.hour == 14]

fig, ax = plt.subplots()
for t in pdata.Time.unique():
    p = pdata[pdata.Time == t]
    dots = ax.scatter(p["Time"].dt.minute, np.full(len(p), 1.0), facecolors='none', edgecolors='r')
    jitter(dots)
ax.tick_params(labelleft=False, left=False)

h = [datetime.datetime(2017, 1, 1, 14, int(min)) for min in ax.get_xticks()[1:-2]]
h.append(datetime.datetime(2017, 1, 1, 15))
h.append(datetime.datetime(2017, 1, 1, 15, 10))
h.insert(0, datetime.datetime(2017, 1, 1, 13, 50))
labels = [t.strftime("%H:%M") for t in h]
ax.set_xticklabels(labels)

plt.xlabel("Time of arrest")
ymin, ymax = plt.ylim()
plt.ylim(ymin - 1.0, ymax + 1.0)
plt.show()
```

> * Also I want to show you guys another interesting use of box plots. To get started, let's calculate another average for each focus-crime, namely what time of day the crime happens. So this time, the distribution we want to plot is the average time-of-day that a crime takes place. There are many ways to do this, but let me describe one way to do it. 
    * For datapoint, the only thing you care about is the time-of-day, so discard everything else.
    * You also have to deal with the fact that time is annoyingly not divided into nice units that go to 100 like many other numbers. I can think of two ways to deal with this.
      * For each time-of-day, simply encode it as seconds since midnight.
      * Or keep each whole hour, and convert the minute/second count to a percentage of an hour. So 10:15 $\rightarrow$ 10.25, 8:40 $\rightarrow$ 8.67, etc.
    * Now you can create box-plots to create an overview of *when various crimes occur*. Note that these plot have quite a different interpretation than ones we created in the previous exercise. Cool, right?  

```python
import scipy as scy
fig, axs = plt.subplots(2, 1, figsize=(10, 8), constrained_layout=True)


scy.stats.probplot(missing.Y, plot=axs[0])
scy.stats.probplot(burglary.Y, plot=axs[1])
axs[0].set_title("Probability plot of missing person arrests")
axs[1].set_title("Probability plot of burglary arrests")
for a in axs:
    a.set_xlabel("Multiples of standard deviation")
    a.set_ylabel("Latitude values")
```

```python
focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])

dists = []
for c in focuscrimes:
    crime_cat = data.loc[data["Category"] == c]
    daily = crime_cat.groupby(pd.Grouper(key="Date", freq="1D")).size()
    dists.append(daily)
fig, ax = plt.subplots(figsize=(18, 9))
ax.boxplot(dists, vert=False, showmeans=True)
ax.set_yticklabels(focuscrimes, fontsize=12)
plt.show()
```

Now the distributions of the focuscrimes can be compared more easily. This plot shows that there is a great number of outliers in the dataset. It also shows that the widest distribution is that of larceny/theft.

```python
dists = []
for c in focuscrimes:
    crime_cat = data.loc[data["Category"] == c]
    hour_fractions = np.around(crime_cat["Time"].dt.hour + crime_cat["Time"].dt.minute / 60 + crime_cat["Time"].dt.second / 3600, decimals=3)
    dists.append(hour_fractions)
fig, ax = plt.subplots(figsize=(18, 9))
ax.boxplot(dists, vert=False, showmeans=True)
ax.set_yticklabels(focuscrimes, fontsize=12)
plt.show()
```

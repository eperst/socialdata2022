```python
### HYPOHESIS TESTING ###

df_female = pd.read_csv('https://raw.githubusercontent.com/suneman/socialdata2022/main/files/data9b_f.csv', sep=',', 
                    header=0, names=["steps", "bmi"])
df_male = pd.read_csv('https://raw.githubusercontent.com/suneman/socialdata2022/main/files/data9b_m.csv', sep=',', 
                    header=0, names=["steps", "bmi"])
```

```python
# H1: there is a statistically significant difference in the average number of steps taken by men and women

#pvalue = h1_result[1]
ttest_result = stats.ttest_ind(df_female['steps'], df_male['steps'])
alpha = 0.05

if ttest_result[1] > alpha:
    print("H1 verified: there IS a statistically significant difference in the average number of steps taken by men and women")
else:
    print("H2 falsified: there is NOT a statistically significant difference in the average number of steps taken by men and women")
    
```

```python
# H2: there is a negative correlation between the number of steps and the BMI for women

pearson_women = np.round(np.corrcoef(df_female["steps"], df_female["bmi"]), 3)
print("Pearson correlation between steps and bmi for women is\n", pearson_women)
print("H2 verified: there IS a negative correlation between the number of steps and the BMI for women")
```

```python
# H3: there is a positive correlation between the number of steps and the BMI for men.

pearson_men = np.round(np.corrcoef(df_male["steps"], df_male["bmi"]), 3)
print("Pearson correlation between steps and bmi for men is\n", pearson_men)
print("H3 falsified: there is NOT a positive correlation between the number of steps and the BMI for men.")
```

**Summary statistics:** Mean, median and related summary statistics work very well for analyzing data if the data set fulfills certain conditions, like being symmetric and not having significant outliers. However, many data sets do not fulfill these conditions, and this can lead to misleading summary statistics. (give example?) By visualizing the full data we can clearly see important properties of the data, for instance if we have outliers. If we want to, we can also calculate summary statistics, but then we have the properties of the data set in mind and won't be fooled.

**Box plots:** A box plot is a graphic which is used to present the median and percentiles of a data set in an easy way. It also displays the outliers. A box plot makes it easy to see if the data is symmetric and the weight distribution of the dataset. It's very useful for getting a clear overview of some important statistics for the data, but the catch is that we lose much important information about the data. For instance, if the data has multiple peaks it can be easily seen in a histogram, but it's not visible in a box plot

A box plot is most useful when we want to compare distributions against each other, since they are very easy to separate and it's easy to see the difference. Using a KDE (Kernel Density Estimation) or a histogram will provide more information, but this also leads to it being more difficult to separate and compare the distributions.

A violin plot is similar to a box plot but it is more informative since it includes a rotated kernel density plot. This means that in addition to showing summary statistics, it also shows the complete distribution of the data. Nevertheless, the violin plot is a less popular choice than box plots, and for people who are not familiar with violin plots it can be difficult to understand the meaning of it. Using simple box plots or KDEs can therefore be a better choice to ensure that the plot is accessible for most people, as it's more important that the plot is understood correctly than including all the information. However, for people who are familiar with violin plots, it might be a better solution than box plots as it shows the information we have in a normal box plot but adds additional information about the distribution.

```python
# BOX AND WHISKERS PLOTS: TIME-OF-DAY DISTRIBUTION OF CRIMES PER CATEGORY

# Function that takes in times of arrest of a crime category and converts them to
# number of hours, rounded to 3 decimal places
def time_as_hours(crime_cat):
    hours = crime_cat.dt.hour 
    minutes = crime_cat.dt.minute / 60
    seconds = crime_cat.dt.second / 3600
    return np.around(hours + minutes + seconds, decimals=3)

# Creates a list from the arrest time distributions of the focus categories
dists = []
for c in focuscrimes:
    crime_cat = filtered.loc[filtered["Category"] == c]
    hour_fractions = time_as_hours(crime_cat["Time"])
    dists.append(hour_fractions)



fig, ax = plt.subplots(figsize=(20, 14))
ax.boxplot(dists, vert=False, showmeans=True, labels=focuscrimes)
# Set hourly label frequency on the x axis
ax.set_xticks(np.arange(0, 25, 1))

# Set font size, figure title and axis labels
ax.tick_params(labelsize=20)
ax.set_title("Time of day a crime occurs within each focus-crime category", fontsize=24)
ax.set_xlabel("Time of day", fontsize=20)
ax.set_ylabel("Category", fontsize=20)

plt.show()
```

# 4b: Heatmap Movie

```python
crimes_in_time_period = burglary_arrests.loc[(burglary_arrests['Date_Time'] >= pd.Timestamp("2005-01-01")) & (burglary_arrests['Date_Time'] <= pd.Timestamp("2006-12-31"))]
crimes_in_time_period.sort_values(by=['Date_Time'], inplace=True, ascending=True)
groupByTime = crimes_in_time_period.groupby(pd.Grouper(key='Date_Time', freq='M'))
xVals = groupByTime['X'].apply(list)
yVals = groupByTime['Y'].apply(list)

frames = []
for i in range(len(xVals)):
    frames.append([[y, x] for y, x in zip(yVals[i], xVals[i])])
SFmap = folium.Map([37.7749, -122.4194], zoom_start=13)
folium.Marker([37.77919, -122.41914], popup='SF City Hall').add_to(SFmap)
HeatMapWithTime(frames).add_to(SFmap)

SFmap
```

The above heatmap movie shows the combination of temporal and spatial data over a two year basis, separated by months.
Interestingly, it illustrates that certain neighborhoods have different peak months for burglaries (i.e. they are not evenly distributed). This indicates that when there is one burglary in a neighborhood, there are likely to have been others in the same month.

#### Bias and misconceptions introduced by the mentioned data-errors ####
1. There is a surprisingly high amount of arrests in a small time period on Thursdays, as shown in the figure above. This is a significant outlier, as the number of arrests here is higher than in all other daily maximum points. This is highly unlikely to occur, and is due to a data duplication errror. This error would be more difficult to notice in the aggregated or binned version of the data, which would be misleading.


2. Rounding the recorded arrest times to the nearest hour, or for example to the nearest 10 minute increment, can give the false impression of a regular, repeating pattern of crimes. Based on our dataset, one could expect more crimes to occur at the beinning of every hour. This assumption is incorrect and stems from the mentioned data-errors. The binned version of such data will also show a false distribution if bin sizes that correspond to the rounding errors are not chosen.


3. This data-error introduces a false hotspot into the data and can hider the discovery of relevant spatial patterns. Such errors can also lead to incorrect predictions and can ultimately cause officers to be dipatched to the wrong location.

```python
import datetime

def jitter(dots):
    offsets = dots.get_offsets()
    offsets[:, 1] += np.round(np.random.uniform(-0.4, 0.4, offsets.shape[0]), decimals=2)
    dots.set_offsets(offsets)

assaults_jan_2017 = filtered.loc[(filtered["Category"] == "ASSAULT") & (filtered["Date"] >= '2017-01-01') & (filtered["Date"] <= '2017-01-31')]
pdata = assaults_jan_2017.loc[assaults_jan_2017.Time.dt.hour == 14]

fig, ax = plt.subplots()
for t in pdata.Time.unique():
    p = pdata[pdata.Time == t]
    dots = ax.scatter(p["Time"].dt.minute, np.full(len(p), 1.0), facecolors='none', edgecolors='r')
    jitter(dots)
ax.tick_params(labelleft=False, left=False)

h = [datetime.datetime(2017, 1, 1, 14, int(min)) for min in ax.get_xticks()[1:-2]]
h.append(datetime.datetime(2017, 1, 1, 15))
h.append(datetime.datetime(2017, 1, 1, 15, 10))
h.insert(0, datetime.datetime(2017, 1, 1, 13, 50))
labels = [t.strftime("%H:%M") for t in h]
ax.set_xticklabels(labels)
plt.suptitle("Times of arrest in the assault category in January 2017 from 14:00 to 15:00")
plt.xlabel("Time of arrest")
ymin, ymax = plt.ylim()
plt.ylim(ymin - 1.0, ymax + 1.0)
plt.show()
```

The above jitter plot displays the times of arrest in the assault category between 14:00 and 15:00 in January of 2017. It is easy to notice that most of the arrests sem to be made on the hour or at regular intervals of 10 minutes. These are errors in the dataset, similar to those in the prostitution category. This means that the arrest time rounding error is present in other categories as well.
